```python
# Import pustaka yang diperlukan
import pandas as pd

# Path file CSV
file_path = "data_kasus_dbd_dki_jakarta_2015_2020.csv"  # Ganti dengan nama file CSV Anda
output_path = "DataOri.csv"  # Nama file untuk hasil preprocessing

# Membaca file CSV
try:
    data = pd.read_csv(file_path)
    print("File berhasil dibaca.")

    # Menampilkan lima baris pertama dari data awal
    print("Data awal:")
    print(data.head())

    # Memastikan kolom 'periode' dan 'jumlah_kasus' ada
    if 'periode' in data.columns and 'jumlah_kasus' in data.columns:
        # Memisahkan kolom 'periode' dan 'jumlah_kasus'
        df_preprocessed = data[['periode', 'jumlah_kasus']].copy()
        
        # Menyimpan hasil ke file baru
        df_preprocessed.to_csv(output_path, index=False)
        print(f"\nData setelah preprocessing berhasil disimpan ke {output_path}.")
    else:
        print("Kolom 'periode' atau 'jumlah_kasus' tidak ditemukan dalam data.")
except FileNotFoundError:
    print(f"File {file_path} tidak ditemukan. Pastikan file ada di direktori yang sesuai.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import KFold
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

# Load Data
data = pd.read_csv( r"DataOri.csv")  # Ganti dengan path dataset Anda
data["periode"] = pd.to_datetime(data["periode"], format="%Y-%m")
data.set_index("periode", inplace=True)

# Normalisasi Data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data[["jumlah_kasus"]])

# Fungsi untuk Membuat Windowed Dataset
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

# Hyperparameter
SEQ_LEN = 12  # Menggunakan data 12 bulan sebelumnya untuk prediksi
BATCH_SIZE = 16
EPOCHS = 50
DROPOUT_RATE = 0.2

# Siapkan Dataset
X, y = create_sequences(data_scaled, SEQ_LEN)

# Split Data untuk Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
fold = 1

# Simpan Evaluasi untuk Setiap Fold
evaluation_results = []

for train_index, test_index in kf.split(X):
    print(f"Fold {fold}:")
    
    # Train-Test Split
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Model Arsitektur
    model = Sequential([
        LSTM(64, activation="relu", return_sequences=True, input_shape=(SEQ_LEN, 1)),
        Dropout(DROPOUT_RATE),
        LSTM(32, activation="relu"),
        Dropout(DROPOUT_RATE),
        Dense(1)
    ])
    
    model.compile(optimizer="adam", loss="mean_squared_error")
    
    # Pelatihan Model
    history = model.fit(X_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1, validation_data=(X_test, y_test))
    
    # Prediksi
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)
    actuals = scaler.inverse_transform(y_test.reshape(-1, 1))
    
    # Evaluasi
    mse = mean_squared_error(actuals, predictions)
    rmse = mse ** 0.5
    mae = mean_absolute_error(actuals, predictions)
    r2 = r2_score(actuals, predictions)
    mape = np.mean(np.abs((actuals - predictions) / actuals)) * 100  # Evaluasi MAPE
    
    evaluation_results.append((mse, rmse, mae, r2, mape))
    print(f"MSE: {mse:.4f}, RMSE: {rmse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}, MAPE: {mape:.4f}%")
    fold += 1

# Visualisasi Hasil Fold Terakhir
plt.figure(figsize=(12, 6))
plt.plot(actuals, label="Actual")
plt.plot(predictions, label="Predicted")
plt.title("Prediksi vs Data Aktual (Fold Terakhir)")
plt.xlabel("Waktu")
plt.ylabel("Jumlah Kasus")
plt.legend()
plt.show()

# Menampilkan Tabel Data Prediksi dan Aktual
results_table = pd.DataFrame({
    "Actual": actuals.flatten(),
    "Predicted": predictions.flatten()
})
print("\nTabel Data Prediksi vs Aktual (Fold Terakhir):")
print(results_table.head(10))  # Tampilkan 10 baris pertama

# Menyimpan Tabel ke CSV (opsional)
results_table.to_csv("prediksi_vs_aktual_fold_terakhir.csv", index=False)

# Rata-rata Evaluasi Cross-Validation
results_df = pd.DataFrame(evaluation_results, columns=["MSE", "RMSE", "MAE", "R2", "MAPE"])
print("\nHasil Rata-Rata Cross-Validation:")
print(results_df.mean())

# Menampilkan Evaluasi dalam Bentuk Tabel
print("\nTabel Evaluasi Cross-Validation untuk Semua Fold:")
print(results_df)
```

```python
import pandas as pd
import numpy as np
from keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Muat model dari file .h5
fold = 6  # Ganti dengan nilai fold yang sesuai
model_path = f"lstm_model_fold_{fold}.h5"  # Ganti dengan path model yang sesuai
model = load_model(model_path)

# Tampilkan ringkasan model
model.summary()

# Muat data dari file CSV
csv_file = 'DataOri.csv'  # Ganti dengan path file CSV
df = pd.read_csv(csv_file)

# Periksa apakah data ada dan sudah diproses dengan benar
print(df.head())  # Menampilkan beberapa baris pertama dari data

# Misalnya, kita ingin memprediksi berdasarkan kolom 'jumlah_kasus'
data = df["jumlah_kasus"].values  # Ambil data target atau fitur yang digunakan
data = data.reshape(-1, 1)  # Ubah dimensi data menjadi 2D untuk scaler

# Normalisasi data (pastikan scaler yang digunakan adalah scaler yang digunakan saat pelatihan)
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Membentuk input sequence untuk prediksi (misalnya, data yang dipakai adalah time-series)
# Sesuaikan ukuran window dengan yang digunakan saat pelatihan
sequence_length = 12  # Contoh: 12 bulan untuk memprediksi bulan ke-13
X = []
for i in range(len(data_scaled) - sequence_length):
    X.append(data_scaled[i:i+sequence_length])

X = np.array(X)

# Lakukan prediksi
y_pred = model.predict(X)

# Invers transformasi untuk mengembalikan ke skala asli
y_pred_original_scale = scaler.inverse_transform(y_pred)

# Tampilkan hasil prediksi
print("Prediksi:", y_pred_original_scale[:12])  # Menampilkan 12 bulan prediksi

# Visualisasi hasil prediksi
plt.plot(df["jumlah_kasus"].values, label="Data Asli")
plt.plot(np.arange(len(df), len(df) + 12), y_pred_original_scale[:12], label="Prediksi")
plt.legend()
plt.title("Prediksi Kasus Menggunakan Model LSTM")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Kasus")
plt.show()
```

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import load_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_error as rmse
from sklearn.preprocessing import StandardScaler
import numpy as np

# Fungsi untuk memuat model ARIMA
def load_arima_model(model_path):
    with open(model_path, 'rb') as file:
        arima_model = pickle.load(file)
    return arima_model

# Fungsi untuk prediksi menggunakan model Transformer dengan pemrosesan data yang benar
def transformer_prediction(model, df):
    # Ambil kolom yang relevan (misalnya 'jumlah_kasus' untuk prediksi)
    features = df[['jumlah_kasus']]  # Sesuaikan jika Anda menggunakan fitur lainnya
    
    # Lakukan normalisasi data (jika model Anda dilatih dengan data yang dinormalisasi)
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    
    # Ubah bentuk data menjadi (samples, timesteps, features) yang diharapkan oleh model
    timesteps = 1  # Sesuaikan jika model Anda menggunakan window waktu tertentu (misalnya, beberapa bulan sebelumnya)
    X = np.reshape(features_scaled, (features_scaled.shape[0], timesteps, features_scaled.shape[1]))
    
    # Lakukan prediksi menggunakan model Transformer
    y_pred = model.predict(X)
    
    return y_pred

# Memuat model ARIMA
arima_model = load_arima_model('sarimax_model.pkl')  # Sesuaikan dengan path model ARIMA

# Memuat model Transformer
fold = 6  # Sesuaikan nomor fold jika perlu
transformer_model = load_model(f"lstm_model_fold_{fold}.h5")  # Sesuaikan dengan path model Transformer

# Memuat dataset (format CSV)
df = pd.read_csv('DataOri.csv')  # Sesuaikan dengan path dataset Anda

# Melakukan prediksi menggunakan model Transformer
y_pred_transformer = transformer_prediction(transformer_model, df)

# Mengecek jika model ARIMA adalah dictionary dan mendapatkan prediksi
if isinstance(arima_model, dict):
    # Pastikan model ARIMA dalam bentuk hasil fitting
    y_pred_arima = arima_model['model'].get_forecast(steps=len(df)).predicted_mean
else:
    # Jika model sudah benar, langsung prediksi
    y_pred_arima = arima_model.predict(n_periods=len(df))  # Sesuaikan dengan panjang data time series Anda

# Mengambil nilai sebenarnya dari dataset
y_true = df["jumlah_kasus"].tolist()

# Membuat DataFrame untuk perbandingan
comparison_df = pd.DataFrame({
    'Month': range(1, len(df) + 1),
    'Actual': y_true,
    'Transformer Prediction': y_pred_transformer[:len(df)].flatten(),  # Pastikan bentuk array datanya tepat
    'ARIMA Prediction': y_pred_arima.flatten()  # Pastikan bentuk array datanya tepat
})

# Menampilkan tabel perbandingan
print(comparison_df)

# Plotting perbandingan
plt.figure(figsize=(12, 6))
plt.plot(comparison_df['Month'], comparison_df['Actual'], label='Actual', color='blue', linewidth=2)
plt.plot(comparison_df['Month'], comparison_df['Transformer Prediction'], label='Transformer Prediction', color='red', linestyle='--')
plt.plot(comparison_df['Month'], comparison_df['ARIMA Prediction'], label='ARIMA Prediction', color='green', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Jumlah Kasus')
plt.title('Perbandingan Nilai Sebenarnya dan Prediksi')
plt.legend()
plt.grid(True)
plt.show()

# Evaluasi model menggunakan metrik MAE dan RMSE
mae_transformer = mean_absolute_error(y_true, y_pred_transformer[:len(df)].flatten())  # Flatten untuk memastikan bentuk yang tepat
rmse_transformer = np.sqrt(mean_squared_error(y_true, y_pred_transformer[:len(df)].flatten()))

mae_arima = mean_absolute_error(y_true, y_pred_arima.flatten())  # Flatten untuk memastikan bentuk yang tepat
rmse_arima = np.sqrt(mean_squared_error(y_true, y_pred_arima.flatten()))

print(f"MAE Model Transformer: {mae_transformer:.4f}, RMSE: {rmse_transformer:.4f}")
print(f"MAE Model ARIMA: {mae_arima:.4f}, RMSE: {rmse_arima:.4f}")
```
